```python
import os
import numpy as np
import matplotlib.pyplot as plt

path = os.path.abspath(os.path.join(os.getcwd()))
file_path = os.path.join(path, 'step02', 'height.txt')

xs = np.loadtxt(file_path)

# plt.hist(xs, bins='auto', density=True)
# plt.xlabel('Height(cm)')
# plt.ylabel('Probability Density')
# plt.show()

# 平均
```

```python
mu = np.mean(xs)
# 標準偏差
sigma = np.std(xs)

# print(mu, sigma)
```

```python
# 正規分布
def normal(x, mu=0, sigma=1):
  y = 1 / (np.sqrt(2 * np.pi) * sigma) * np.exp(-(x - mu)**2 / (2 * sigma**2))
  return y

np.random.normal(mu, sigma, 5)

x = np.linspace(150, 190, 1000)
y = normal(x, mu, sigma)

plt.hist(xs, bins='auto', density=True)
# 曲線をプロット
plt.plot(x, y)
plt.xlabel('Height(cm)')
plt.ylabel('Probability Density')
plt.show()
```

```python
samples = np.random.normal(mu, sigma, 100000)

plt.hist(xs, bins='auto', density=True, alpha=0.7, label='original')
plt.hist(samples, bins='auto', density=True, alpha=0.7, label='generated')
plt.xlabel('Height(cm)')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```
